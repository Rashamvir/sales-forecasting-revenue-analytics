```python
import pandas as pd
import numpy as np
```

```python
orders = pd.read_csv("olist_orders_dataset.csv")
orders.head()
```

```python
order_items = pd.read_csv("olist_order_items_dataset.csv")
order_items.head()
```

```python
products = pd.read_csv("olist_products_dataset.csv")
products.head()
```

```python
customers = pd.read_csv("olist_customers_dataset.csv")
customers.head()
```

```python
category_translation = pd.read_csv("product_category_name_translation.csv")
category_translation.head()
```

```python
orders_filtered = orders[
    (orders["order_status"] == "delivered") &
    (orders["order_delivered_customer_date"].notna())
].copy()
```

```python
orders_filtered["order_delivered_customer_date"] = pd.to_datetime(
    orders_filtered["order_delivered_customer_date"]
)
```

```python
sales = orders_filtered.merge(
    order_items,
    on="order_id",
    how="inner"
)
```

```python
sales = orders_filtered.merge(
    order_items,
    on="order_id",
    how="inner"
)
```

```python
sales = sales.merge(
    products[["product_id", "product_category_name"]],
    on="product_id",
    how="left"
)
```

```python
sales = sales.merge(
    category_translation,
    on="product_category_name",
    how="left"
)

sales.rename(columns={
    "product_category_name_english": "product_category"
}, inplace=True)
```

```python
sales = sales.merge(
    customers[["customer_id", "customer_state", "customer_city"]],
    on="customer_id",
    how="left"
)
```

```python
sales["total_item_revenue"] = sales["price"] + sales["freight_value"]
```

```python
sales["year"] = sales["order_delivered_customer_date"].dt.year
sales["month"] = sales["order_delivered_customer_date"].dt.month
sales["year_month"] = sales["order_delivered_customer_date"].dt.to_period("M").astype(str)
```

```python
final_df = sales[
    [
        "order_id",
        "order_delivered_customer_date",
        "year",
        "month",
        "year_month",
        "product_id",
        "product_category",
        "seller_id",
        "customer_state",
        "customer_city",
        "price",
        "freight_value",
        "total_item_revenue"
    ]
].copy()
```

```python
final_df.to_csv(
    "sales_revenue_analytics.csv",
    index=False
)
```
