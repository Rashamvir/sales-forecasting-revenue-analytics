```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sales_revenue_analytics.csv")
df["order_delivered_customer_date"] = pd.to_datetime(
    df["order_delivered_customer_date"]
)
```

```python
monthly_revenue = (
    df.groupby("year_month")["total_item_revenue"]
    .sum()
    .reset_index()
)

monthly_revenue["year_month"] = pd.to_datetime(monthly_revenue["year_month"])
```

```python
plt.figure()
plt.plot(
    monthly_revenue["year_month"],
    monthly_revenue["total_item_revenue"]
)
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
monthly_orders = (
    df.groupby("year_month")["order_id"]
    .nunique()
    .reset_index(name="order_count")
)

monthly_orders["year_month"] = pd.to_datetime(monthly_orders["year_month"])

plt.figure()
plt.plot(
    monthly_orders["year_month"],
    monthly_orders["order_count"]
)
plt.title("Monthly Order Volume Trend")
plt.xlabel("Month")
plt.ylabel("Orders")
plt.show()
```

```python
monthly_aov = monthly_revenue.merge(
    monthly_orders,
    on="year_month"
)

monthly_aov["aov"] = (
    monthly_aov["total_item_revenue"] / monthly_aov["order_count"]
)

plt.figure()
plt.plot(
    monthly_aov["year_month"],
    monthly_aov["aov"]
)
plt.title("Average Order Value (AOV) Trend")
plt.xlabel("Month")
plt.ylabel("AOV")
plt.show()
```

```python
seasonality = (
    df.groupby("month")["total_item_revenue"]
    .sum()
    .reset_index()
)

plt.figure()
plt.bar(seasonality["month"], seasonality["total_item_revenue"])
plt.title("Revenue Seasonality by Month")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.show()
```

```python
top_categories = (
    df.groupby("product_category")["total_item_revenue"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_categories
```
